## Scraping Data of Countries in the world by population From worldometers :
### URL : https://www.worldometers.info/world-population/population-by-country/

```python
import requests, re
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from IPython.display import display, HTML

import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"

response = requests.get(url)

# Parsing the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'id': 'example2'})
# display(HTML(str(table)))
```

```python
# Function to clean and convert numbers
def clean_number(number_str):
     # Remove commas and any non-numeric characters
    cleaned_str = re.sub(r'[^0-9]', '', number_str)
    if cleaned_str :
        return int(cleaned_str)
    return cleaned_str
```

```python
table_content = []
all_lines = table.find('tbody')
all_lines = table.find_all('tr')

for line in all_lines :
    line_info = line.find_all('td')
    if line_info :
        table_content.append({
            'id': clean_number(line_info[0].text.strip()),
            'Country': line_info[1].text.strip(),
            'Population': clean_number(line_info[2].text.strip()), 
            'Yearly Change (%)': line_info[3].text.strip(), 
            'Net Change': clean_number(line_info[4].text.strip()),
            'Density (P/Km²)': clean_number(line_info[5].text.strip()), 
            'Land Area (Km²)': clean_number(line_info[6].text.strip()), 
            'Migrants (net)': clean_number(line_info[7].text.strip()),
            'Fertility Rate': line_info[8].text.strip(), 
            'Mid Age': clean_number(line_info[9].text.strip()),
            'Urban Pop (%)': line_info[10].text.strip(),
            'World Share (%)': line_info[11].text.strip(),
        })
```

```python
data = pd.DataFrame(table_content)
data
```
