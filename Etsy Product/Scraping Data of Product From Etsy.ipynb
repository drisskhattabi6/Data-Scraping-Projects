# Scraping Data of Product From Etsy :

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# url = "https://www.etsy.com/listing/1473301320/luxury-tournament-chess-set-large-chess"
# url = "https://www.etsy.com/listing/1217927127/luxury-chess-set-handmade-chess-board"
# url = "https://www.etsy.com/listing/1163727616/best-friends-necklace-for-two-split"
url = "https://www.etsy.com/listing/1670733128/pixelated-night-light-lantern-chargeable"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Cache-Control': 'max-age=0'
}

response = requests.get(url, headers=headers)

# Parsing the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

product_info = soup.find('div', {'id': 'listing-page-cart'})
#---------------------------------------
price = product_info.find('div', {'data-buy-box-region': 'price'}).p

span_tag = price.find('span', class_='wt-screen-reader-only')

# Remove the <span> tag
if span_tag:
    span_tag.decompose()

price = price.text.strip()
#---------------------------------------
title = product_info.find('h1', {'data-buy-box-listing-title': 'true', 'class': 'wt-line-height-tight wt-break-word wt-text-body-01'}).text.strip()
#---------------------------------------
rating = product_info.find('input', {'name': 'rating', 'type': 'hidden'}).attrs['value']
#---------------------------------------
product_details = soup.find('div', {'id': 'product_details'})
#---------------------------------------
made_by = product_details.find('ul', {'data-selector': 'product-details-highlights'})
made_by = made_by.find_all('li')[0]
made_by = made_by.find('a').text.strip()
#---------------------------------------
description = product_details.find('div', {'data-id': 'description-text'})

# Replace <br> tags with newline characters
for br in description.find_all('br'):
    br.replace_with('\n')

description = description.p.text.strip()
#---------------------------------------
shipping_container = soup.find('div', {'id': 'shipping-and-returns-div'})
shipping_info = shipping_container.find('div', {'data-selector': 'shipping-highlights'}).ul
shipping_info = shipping_info.find_all('li')
#---------------------------------------
expicted_date = shipping_info[0].strong
if expicted_date :
    expicted_date = expicted_date.text.strip()
#---------------------------------------
cost_to_ship = shipping_info[2].strong.text.strip()
#---------------------------------------
ships_from = shipping_info[3].strong.text.strip()
#---------------------------------------
print('price : ', price)
print('*'*50)
print('title : ', title)
print('*'*50)
print('rating : ', rating)
print('*'*50)
print('made_by : ', made_by)
print('*'*50)
print('description : ', description)
print('*'*50)
print('cost_to_ship : ', cost_to_ship)
print('*'*50)
print('expicted_date : ', expicted_date)
print('*'*50)
print('ships_from : ', ships_from)
```
