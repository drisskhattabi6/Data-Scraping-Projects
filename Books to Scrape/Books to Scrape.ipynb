# Scraping Books from [books toscrape website](https://books.toscrape.com/)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
all_books_list = []

for nbr in range(1, 51) :

    url = f'https://books.toscrape.com/catalogue/page-{nbr}.html'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    all_books = soup.find('ol', {'class' : 'row'})
    all_books = all_books.find_all('li')

    # print(len(all_books))
    for book in all_books :
        data = {}

        article = book.find('article')
        book_info = article.find('h3')
        book_link = book_info.find('a').get('href')
        data['book_link'] = f'https://books.toscrape.com/catalogue/{book_link}'
        # print(f'https://books.toscrape.com/catalogue/{book_link}')
        # -----------------------------------------
        book_page = requests.get(f'https://books.toscrape.com/catalogue/{book_link}' , book_link)
        book_page = BeautifulSoup(book_page.content, 'html.parser')
        book_page_content = book_page.find('div', {'id': 'content_inner'})
        # -----------------------------------------
        header_info = book_page_content.find('div', {'class': 'row'})
        book_img_url = header_info.find('img').get('src')
        book_title = header_info.find('h1').text.strip()
        book_price = header_info.find('p', {'class': 'price_color'}).text.strip()
        data['book_img_url'] = book_img_url
        data['book_title'] = book_title
        data['book_price'] = book_price
        # print(book_title)
        # -----------------------------------------
        product_description_div = book_page_content.find('div', {'id':'product_description'})
        product_description_paragraph = None
        if product_description_div != None :
            product_description_paragraph = product_description_div.find_next('p').text.strip()
        data['book_description'] = product_description_paragraph
        # -----------------------------------------

        table = book_page_content.find('table', {'class': 'table table-striped'})
        for row in table.find_all('tr'):
            # Get the header (th) and corresponding data (td)
            header = row.find('th').text.strip()
            value = row.find('td').text.strip()
            data[header] = value

        all_books_list.append(data)

#         for key, value in data.items():
#             print(f"{key}: {value}")

    print('nbr -> ', nbr)
```

```python
data = pd.DataFrame(all_books_list)
data.head()
```
