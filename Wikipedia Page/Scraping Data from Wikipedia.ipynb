# Scraping Data from Wikipedia :

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, HTML
```

```python
def get_core_page(url) :
    response = requests.get(url)
    
    # Parsing the webpage content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    article_container = soup.find('main', {'id': 'content'})

    article_title = article_container.find('h1', {'id': 'firstHeading'})
    article_content = article_container.find('div', {'id': 'mw-content-text'})

    return article_title, article_content
    
```

```python
url = "https://en.wikipedia.org/wiki/Genetic_algorithm"
title, content = get_core_page(url)
display(HTML(str(title)))
display(HTML(str(content)))
```
