# Scraping Data of Hockey Teams From [scrape this site](scrapethissite.com) :
#### from : https://www.scrapethissite.com/pages/forms/

```python
import requests, re, csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def text_processing(text):
    if type(text) == str :
        # pattern to match [, \n, and ]
        pattern = r'[\[\]\n]'
        
        # Replace the unwanted characters with an empty string
        cleaned_text = re.sub(pattern, '', text)
    
        if cleaned_text :
            cleaned_text = cleaned_text.strip()
        
        return cleaned_text
    return text
```

```python
#----------------------
page_nbrs = 3
#----------------------
table_content = []
line_id = 1

for page_nbr in range(1, page_nbrs+1) :
    
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page_nbr}"
    
    response = requests.get(url)
        
    # Parsing the webpage content
    soup = BeautifulSoup(response.content, 'html.parser')
        
    table = soup.find('table', {'class': 'table'})
        
    all_lines = table.find('tbody')
    all_lines = table.find_all('tr', {'class': 'team'})

    for line in all_lines :
        table_content.append({
            'Id': line_id, 
            'Team Name': text_processing(line.find('td', {'class': 'name'}).text), 
            'Year': text_processing(line.find('td', {'class': 'year'}).text), 
            'Wins': text_processing(line.find('td', {'class': 'wins'}).text), 
            'Losses': text_processing(line.find('td', {'class': 'losses'}).text),
            'OT Losses': text_processing(line.find('td', {'class': 'ot-losses'}).text), 
            'Win %': text_processing(line.find('td', {'class': 'pct'}).text), 
            'Goals For (GF)': text_processing(line.find('td', {'class': 'gf'}).text),
            'Goals Against (GA)': text_processing(line.find('td', {'class': 'ga'}).text), 
            '+ / -': text_processing(line.find('td', {'class': 'diff'}).text)
        })
        line_id += 1
```

### storing the scraped data in a CSV file

```python
# first method
pd.DataFrame(table_content).to_csv(path_or_buf="Hockey_Teams.csv")
```
